```python
def Delc_bypair_plot(osc1, osc2, osc3, osc4, osc5, xlim, ylim, ytick, osc_type, save): 
    
    mpl.rcParams.update({'font.size': 12})
    
    plt.style.use('/Users/clay/Documents/research/jgr20/summary_data/jgr2020_style2.mplstyle')
    ax_clr = np.array([[0, 0.4470, 0.7410], [0.9290, 0.6940, 0.1250], [0.4940, 0.1840, 0.5560], [0.8500, 0.3250, 0.0980], [0.4660, 0.6740, 0.1880]])

    if osc_type == 'NS':
        clr1 = '#B8221E'
        clr2 = '#fdb366'
    else:
        clr1 = '#2166AC'
        clr2 = '#92C5DE'

    freqs = np.array([0.1,1,10,40])
    mrkr = np.array(['d','o','*','s'])
    mrkrsz = np.array([6,6,9,5])

    order = np.array([3, 0, 1, 2])+3; 
    
    fig, axs = plt.subplots(4, 3, figsize=(10,12), sharex=True, sharey=True, gridspec_kw={'hspace': 0.09, 'wspace':0.11})
#     gridspec_kw={'hspace': 0.15, 'wspace':0.15}
#     fig.tight_layout()
    
    for aa in range(4): 
#         fig, ax = plt.subplots(1,1,figsize=(3,3),dpi=300)
        
        for bb in range(len(osc1)):
            axs[aa,0].plot(osc1[bb,2], osc1[bb,order[aa]], marker=mrkr[np.where(osc1[bb,1] == freqs)][0], lw=0, mfc=clr1, mew=0.6, mec='k', markersize=mrkrsz[np.where(osc1[bb,1] == freqs)][0])
        for bb in range(len(osc2)):
            axs[aa,0].plot(osc2[bb,2], osc2[bb,order[aa]], marker=mrkr[np.where(osc2[bb,1] == freqs)][0], lw=0, mfc=clr2, mew=0.6, mec='k', markersize=mrkrsz[np.where(osc2[bb,1] == freqs)][0])
        for bb in range(len(osc3)):
            axs[aa,1].plot(osc3[bb,2], osc3[bb,order[aa]], marker=mrkr[np.where(osc3[bb,1] == freqs)][0], lw=0, mfc=clr1, mew=0.6, mec='k', markersize=mrkrsz[np.where(osc3[bb,1] == freqs)][0])
        for bb in range(len(osc4)):
            axs[aa,1].plot(osc4[bb,2], osc4[bb,order[aa]], marker=mrkr[np.where(osc4[bb,1] == freqs)][0], lw=0, mfc=clr2, mew=0.6, mec='k', markersize=mrkrsz[np.where(osc4[bb,1] == freqs)][0])
        for bb in range(len(osc5)):
            axs[aa,2].plot(osc5[bb,2], osc5[bb,order[aa]], marker=mrkr[np.where(osc5[bb,1] == freqs)][0], lw=0, mfc=clr1, mew=0.6, mec='k', markersize=mrkrsz[np.where(osc5[bb,1] == freqs)][0])

    for dd in range(3):
        for aa in range(4):
            for spine in ['left','right','top','bottom']:
                axs[aa,dd].spines[spine].set_color(ax_clr[order[aa]-3])

            # AXES: MINOR TICKS, COLORS
            axs[aa,dd].set_xticks(np.arange(-0.2, 2.2, 0.4))
#             axs[aa,dd].set_yticks(ytick)
            axs[aa,dd].set_xlim(xlim)
            axs[aa,dd].set_ylim(ylim)
#             axs[aa,dd].set_xlim(xlim)
            axs[aa,dd].minorticks_on()
            axs[aa,dd].get_xaxis().set_tick_params(direction='in', top=True, bottom=True,which='both', colors=ax_clr[order[aa]-3])
            axs[aa,dd].get_yaxis().set_tick_params(direction='in', top=True, bottom=True,which='both', colors=ax_clr[order[aa]-3])

            # AXES LABELS & COLORS   
            axs[aa,dd].set_xlabel('',color='w')
            axs[aa,dd].ticklabel_format(axis='y',style='sci',scilimits=(-4,2))
            axs[aa,0].set_ylabel('$\Delta c/c_0$ (%)',color=ax_clr[order[aa]-3])

            axs[3,dd].set_xlabel('$Osc.\ Amp.\ (MPa)$',color=ax_clr[order[aa]-3])
            axs[aa,dd].plot([-0.1, 2], [0, 0], color = 'gray',ls = '--',zorder=1) #Horizontal Line
            

    axs[0,0].set_title('$Post-fracture$')
    axs[0,1].set_title('$Post-shear\ 1$')
    axs[0,2].set_title('$Post-shear\ 2$')
    plt.show()

    if save == 'y':
        fig.savefig('/Users/clay/Documents/research/jgr20/manuscript/new_plots/Delc_bypair_all_'+osc_type+'.svg',transparent=True)
        fig.savefig('/Users/clay/Documents/research/jgr20/manuscript/new_plots/Delc_bypair_all_'+osc_type+'.pdf',transparent=True)
```

```python
xlim = [0.0,1.8]; 
ylim = [-0.06, 0.005]; 
ytick = np.arange(-0.06, 0.2, 0.02)
save='y'
Delc_bypair_plot(f_ns_75, f_ns_66, s1_ns_75, s1_ns_66, s2_ns_75, xlim, ylim, ytick, 'NS', save)
```
